```python
import duckdb
import pandas as pd

path = "/home/aoxu/projects/VLS-Benchmark-Dataset/plate-vs/VLS_benchmark/chembl_affinity/uniprot_O00255/O00255_chembl_activities_filtered.parquet"
fixed = path.replace(".parquet", ".fixed.parquet")

# 1) Try reading with DuckDB
df_affinity = duckdb.read_parquet(path).df()
print("Read rows:", len(df_affinity))

# 2) Rewrite to a fresh parquet
duckdb.sql(f"""
COPY (SELECT * FROM read_parquet('{path}'))
TO '{fixed}'
(FORMAT PARQUET);
""")
print("Wrote:", fixed)

# 3) Now try pyarrow on the rewritten file
df_affinity2 = pd.read_parquet(fixed, engine="pyarrow")
print("Re-read rows:", len(df_affinity2))
```

```python
import os, traceback
import pandas as pd

path = "/home/aoxu/projects/VLS-Benchmark-Dataset/plate-vs/VLS_benchmark/chembl_affinity/uniprot_O00255/O00255_chembl_activities_filtered.parquet"

print("File size (bytes):", os.path.getsize(path))

def try_read(label, fn):
    print(f"\n--- {label} ---")
    try:
        df = fn()
        print("OK rows:", len(df))
        print(df.dtypes.head(10))
        return df
    except Exception as e:
        print("FAILED:", repr(e))
        traceback.print_exc(limit=3)
        return None

df1 = try_read("pyarrow default", lambda: pd.read_parquet(path, engine="pyarrow"))
df2 = try_read("pyarrow legacy_dataset", lambda: pd.read_parquet(path, engine="pyarrow", use_legacy_dataset=True, use_threads=False))
df3 = try_read("fastparquet default", lambda: pd.read_parquet(path, engine="fastparquet"))
df4 = try_read("fastparquet pandas_nulls=False", lambda: pd.read_parquet(path, engine="fastparquet", pandas_nulls=False))
```
